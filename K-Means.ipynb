```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Gerar dados fictícios
np.random.seed(42)
n_clientes = 200

# Frequência de compras (número de compras por mês)
frequencia_compras = np.random.randint(1, 20, size=n_clientes)

# Valor gasto (em reais)
valor_gasto = np.random.randint(50, 1000, size=n_clientes)

# Criar um DataFrame
dados = pd.DataFrame({
    'FrequenciaCompras': frequencia_compras,
    'ValorGasto': valor_gasto
})

# Visualizar os dados
display(dados.head())
```

```python
plt.scatter(dados['FrequenciaCompras'], dados['ValorGasto'], c='blue', s=50)
plt.xlabel('Frequência de Compras')
plt.ylabel('Valor Gasto')
plt.title('Comportamento dos Clientes')
plt.show()
```

```python
# definir o número de clusters
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(dados)
dados['Cluster'] = kmeans.labels_

# Visualizar os clusters
plt.scatter(dados['FrequenciaCompras'], dados['ValorGasto'], c=dados['Cluster'], cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.xlabel('Frequência de Compras')
plt.ylabel('Valor Gasto (R$)')
plt.title('Clientes Agrupados por Comportamento de Compra')
plt.show()
```

```python
# Analisar os clusters
for cluster in range(k):
    print(f"Cluster {cluster}:")
    display(dados[dados['Cluster'] == cluster].describe())
    print("\n")
```
