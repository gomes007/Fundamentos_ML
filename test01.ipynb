```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Criando um conjunto de dados de exemplo
np.random.seed(42)
dados = np.random.randint(10, 100, 50)  # Conjunto de 50 nÃºmeros inteiros aleatÃ³rios entre 10 e 100

# Calculando as estatÃ­sticas
media = np.mean(dados)
mediana = np.median(dados)
percentil_25 = np.percentile(dados, 25)
percentil_75 = np.percentile(dados, 75)
quartis = [percentil_25, percentil_75]
decis = [np.percentile(dados, i) for i in range(10, 100, 10)]

# Imprimindo os resultados
print("MÃ©dia:", media)
print("Mediana:", mediana)
print("Percentil 25:", percentil_25)
print("Percentil 75:", percentil_75)
print("Quartis:", quartis)
print("Decis:", decis)

# Visualizando os dados com grÃ¡ficos
plt.figure(figsize=(14, 6))

# Histograma com mÃ©dia, mediana e moda
plt.hist(dados, bins=10, alpha=0.6, color='lightblue', edgecolor='black')
plt.axvline(media, color='red', linestyle='--', label=f'MÃ©dia ({media:.2f})')
plt.axvline(mediana, color='green', linestyle='--', label=f'Mediana ({mediana:.2f})')


# Adicionando as linhas dos percentis
plt.axvline(percentil_25, color='purple', linestyle=':', label=f'Percentil 25 ({percentil_25:.2f})')
plt.axvline(percentil_75, color='orange', linestyle=':', label=f'Percentil 75 ({percentil_75:.2f})')

# PersonalizaÃ§Ã£o do grÃ¡fico
plt.title('DistribuiÃ§Ã£o dos Dados com EstatÃ­sticas Descritivas')
plt.xlabel('Valores')
plt.ylabel('FrequÃªncia')
plt.legend()

plt.show()
```

```python
import numpy as np

# Criando um conjunto de dados de exemplo
dados = np.random.randint(10, 100, 50)  # Conjunto de 50 nÃºmeros inteiros aleatÃ³rios entre 10 e 100

# Calculando as estatÃ­sticas
amplitude = np.max(dados) - np.min(dados)
variancia = np.var(dados, ddof=1)  # ddof=1 para variÃ¢ncia amostral
desvio_padrao = np.std(dados, ddof=1)  # ddof=1 para desvio padrÃ£o amostral
erro_padrao = desvio_padrao / np.sqrt(len(dados))
coeficiente_variacao = (desvio_padrao / np.mean(dados)) * 100

# Imprimindo os resultados
print("Amplitude:", amplitude)
print("VariÃ¢ncia:", variancia)
print("Desvio PadrÃ£o:", desvio_padrao)
print("Erro PadrÃ£o:", erro_padrao)
print("Coeficiente de VariaÃ§Ã£o (%):", coeficiente_variacao)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Criando um conjunto de dados de exemplo
dados = np.random.randint(10, 100, 50)  # Conjunto de 50 nÃºmeros inteiros aleatÃ³rios entre 10 e 100

# Calculando as estatÃ­sticas
amplitude = np.max(dados) - np.min(dados)
variancia = np.var(dados, ddof=1)  # ddof=1 para variÃ¢ncia amostral
desvio_padrao = np.std(dados, ddof=1)  # ddof=1 para desvio padrÃ£o amostral
erro_padrao = desvio_padrao / np.sqrt(len(dados))
coeficiente_variacao = (desvio_padrao / np.mean(dados)) * 100

# Imprimindo os resultados
print("Amplitude:", amplitude)
print("VariÃ¢ncia:", variancia)
print("Desvio PadrÃ£o:", desvio_padrao)
print("Erro PadrÃ£o:", erro_padrao)
print("Coeficiente de VariaÃ§Ã£o (%):", coeficiente_variacao)

# Plotando o histograma
plt.figure(figsize=(14, 6))
plt.hist(dados, bins=10, alpha=0.6, color='lightblue', edgecolor='black')

# Adicionando linhas representando a mÃ©dia e o desvio padrÃ£o
media = np.mean(dados)
plt.axvline(media, color='red', linestyle='--', label=f'MÃ©dia ({media:.2f})')
plt.axvline(media + desvio_padrao, color='orange', linestyle=':', label=f'MÃ©dia + Desvio PadrÃ£o ({media + desvio_padrao:.2f})')
plt.axvline(media - desvio_padrao, color='orange', linestyle=':', label=f'MÃ©dia - Desvio PadrÃ£o ({media - desvio_padrao:.2f})')

# Adicionando legendas e tÃ­tulos
plt.title('Histograma dos Dados com MÃ©dia e Desvio PadrÃ£o')
plt.xlabel('Valores')
plt.ylabel('FrequÃªncia')
plt.legend()

plt.show()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Criando a tabela de contingÃªncia
dados = pd.DataFrame({
    'AÃ§Ã£o': [30, 10],
    'ComÃ©dia': [20, 40]
}, index=['Masculino', 'Feminino'])

# Calculando o qui-quadrado
chi2, p, dof, expected = chi2_contingency(dados)

print("Valor do qui-quadrado:", chi2)
print("p-valor:", p)
```
