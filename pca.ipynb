```python
# Importação das bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Simulando um dataset de clientes do banco
np.random.seed(42)
num_clients = 200

data = pd.DataFrame({
    'Idade': np.random.randint(18, 70, num_clients),
    'Salario_Anual': np.random.randint(20000, 120000, num_clients),
    'Score_Credito': np.random.randint(300, 850, num_clients),
    'Dividas': np.random.randint(0, 50000, num_clients),
    'Investimentos': np.random.randint(0, 100000, num_clients)
})

# Passo 1: Normalização dos dados
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Passo 2: Aplicação do PCA para reduzir para 2 componentes principais
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Passo 3: Criando um novo DataFrame com os componentes principais
df_pca = pd.DataFrame(data_pca, columns=['PC1', 'PC2'])

# Passo 4: Plotando os dados reduzidos
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_pca['PC1'], y=df_pca['PC2'], alpha=0.7)
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.title("Clientes projetados no espaço PCA")
plt.show()

# Passo 5: Explicação da Variância de Cada Componente
explained_variance = pca.explained_variance_ratio_
print(f"Variância explicada pelo PC1: {explained_variance[0]:.2f}")
print(f"Variância explicada pelo PC2: {explained_variance[1]:.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Criando um pequeno conjunto de dados
dados = pd.DataFrame({
    "Idade": [30, 45, 25],
    "Salario_Anual": [50000, 80000, 40000],
    "Score_Credito": [700, 750, 600],
    "Dividas": [10000, 5000, 15000],
    "Investimentos": [20000, 40000, 10000]
})

# Normalizando os dados para evitar distorções
scaler = StandardScaler()
dados_normalizados = scaler.fit_transform(dados)

# Aplicando PCA para reduzir para 2 componentes
pca = PCA(n_components=2)
dados_pca = pca.fit_transform(dados_normalizados)

# Criando um DataFrame com os novos componentes principais
df_pca = pd.DataFrame(dados_pca, columns=["PC1", "PC2"])
df_pca["Cliente"] = ["João", "Maria", "Pedro"]

# Plotando o resultado
plt.figure(figsize=(6, 4))
plt.scatter(df_pca["PC1"], df_pca["PC2"])
for i, txt in enumerate(df_pca["Cliente"]):
    plt.annotate(txt, (df_pca["PC1"][i], df_pca["PC2"][i]), fontsize=12)
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.title("Clientes Projetados em 2D usando PCA")
plt.show()

# Exibir os novos dados transformados
print(df_pca)
```

```python
# Obtendo os pesos (loadings) das variáveis nos componentes principais
pca_loadings = pd.DataFrame(pca.components_, columns=dados.columns, index=["PC1", "PC2"])

# Exibir a matriz de carga como tabela no Jupyter Notebook
from IPython.display import display
display(pca_loadings)

display(dados)
```
